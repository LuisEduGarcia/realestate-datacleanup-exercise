```python
## Aquí: traza la coordenadas de los estados

## PON TU CÓDIGO AQUÍ:
import pandas as pd


# Leer el archivo CSV  y filtrar cinturon SUR

ds = pd.read_csv('assets/real_estate.csv', sep=';')

ds['latitude'] = ds['latitude'].str.replace(',', '.')
ds['longitude'] = ds['longitude'].str.replace(',', '.')

# Convertir las columnas 'latitude' y 'longitude' a numéricas, manejando errores
ds['latitude'] = pd.to_numeric(ds['latitude'])
ds['longitude'] = pd.to_numeric(ds['longitude'])

# Eliminar filas con valores NaN en 'latitude' o 'longitude'
ds = ds.dropna(subset=['latitude', 'longitude'])

# Eliminar las filas donde 'latitude' o 'longitude' sean iguales a 0
ds = ds[(ds['latitude'] != 0) & (ds['longitude'] != 0)]


# Filtrar datos por municipios
municipios = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
data_filtrado = ds[ds['level5'].isin(municipios)]

color_map = {
    "Fuenlabrada": "red",
    "Leganés": "green",
    "Getafe": "blue",
    "Alcorcón": "yellow"
}

# Crear un diccionario con cada propiedad y sus coordenadas 
properties_dict = {}
for index, row in data_filtrado.iterrows():
    properties_dict[row['url_inmueble']] = {
        'latitude': row['latitude'],
        'longitude': row['longitude'],
        'municipio': row['level5'],
        'color': color_map.get(row['level5'])
    }

# Mostrar el diccionario
print(properties_dict)
```

```python
from ipyleaflet import Map, Circle, LayerGroup

# Crear un mapa centrado en un punto central (e.g., el centro geográfico de los datos)

center_lat = data_filtrado['latitude'].mean()
center_lon = data_filtrado['longitude'].mean()

m = Map(center=(center_lat, center_lon), zoom=11)

# Crear un grupo de capas para los círculos
circle_layer = LayerGroup()

# Añadir círculos al mapa
for prop_id, props in properties_dict.items():
    circle = Circle(location=(props['latitude'], props['longitude']), draggable=False, radius=3, color=props['color'], fill_color=props['color'], fill_opacity=0.6)
   
    circle_layer.add_layer(circle)

# Añadir la capa de círculos al mapa
m.add_layer(circle_layer)

# Mostrar el mapa
m


```
